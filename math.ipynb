1. Number Functions


```python
import math
print(math.ceil(2.3))
print(math.floor(2.7))
print(math.trunc(8.9))
```

2. Power and Logarithmic Functions

```python
print(math.pow(2, 3))
print(math.sqrt(16))
print(math.exp(2))
print(math.log(100, 10))
```

3. Trigonometric Functions

```python


print(math.sin(math.pi/2))
print(math.cos(0))
print(math.tan(math.pi/4))

print(math.radians(90))

print(math.degrees(math.pi))


```

4. Special Functions

```python
print(math.factorial(5))
print(math.gcd(20, 8))
print(math.fabs(-7.5))
```

5. Constants

```python
print(math.pi)
print(math.e)
```
