```python
import numpy as np
from statistics import mean, median, mode
array1 = np.array([1,2,3,4,5])
print(array1)
```

```python


# Central Tendency Measures
print("Mean:", mean(array1))         # Mean
print("Median:", median(array1))     # Median
print("Mode:", mode(array1))         # Mode

# Measures of Dispersion
print("Variance:", np.var(array1))            # Variance (Population)
print("Standard Deviation:", np.std(array1))
```

```python
import numpy as np

my_array = np.array([1,2,3,4])

# Print `my_array` before appending
print("my_array before appending:", my_array)

# Append a 1D array to `my_array`
new_array = np.append(my_array, [7, 8, 9, 10])

# Print `new_array`
print("new_array:", new_array)

# Print `my_array` after appending
print("my_array after appending:", my_array)
```
